# Order Issue Detector

**Purpose:** Quickly identify “abnormal” orders that require attention (e.g., order split, pending shipment, credit hold) from SAP order exports where issues are not always clearly flagged.

**Why this exists:** SAP is comprehensive, but even in Excel it’s hard to spot the few actionable problem orders. Some orders should be held or escalated but may not be surfaced in time, so this notebook acts as a practical triage layer.

**Input:** SAP order export (Excel/CSV).

**How to use:** Download the latest file → Run

**What it outputs:**
- Filtered list of candidate abnormal orders (split-required / waiting-to-ship / credit-hold-related)
- Reason tags / conditions that triggered each flag (for quick review)

**Outcome:** Enables fast escalation and reduces the risk of missed holds, delayed shipments, or partial fulfillment surprises.

```python
import pandas as pd
import openpyxl
import numpy as np
from openpyxl import load_workbook
from pandas import DataFrame
```

Input the raw SAP data file path below

```python
raw=pd.read_excel("C:\\Users\\zhuoyan.bai\\Documents\\DO\\dilvery # line split.xlsx",converters={'Order #':str,'P/O #':str,'Delivery #':str,'Item #':str})
```

```python
deli=[]
split={'Pet':[],'SKU':[],'SO':[],'PO':[],'Plant':[],'Name':[],'ToSt':[],'Qty':[],'Line':[],'SLOC':[],'Comment':[]}
for i in enumerate(raw['Order #']):
    try:
        cat=i[1]+' '+raw.loc[i[0],'Item #']
        if pd.notnull(raw.loc[i[0],'Delivery #']):
            if cat in deli: 
                continue
            else:
                #cat=cat+' '+raw.loc[i[0],'Delivery #']
                deli.append(cat)
        elif pd.isnull(raw.loc[i[0],'Delivery #']) and cat in deli:
            split['Pet'].append(str(raw.loc[i[0],'Old Mat. #']))
            split['SKU'].append(str(raw.loc[i[0],'Model #']))
            split['SO'].append(str(raw.loc[i[0],'Order #']))
            split['PO'].append(str(raw.loc[i[0],'P/O #']))
            split['Plant'].append(str(raw.loc[i[0],'Plant']))
            split['Name'].append(str(raw.loc[i[0],'Name']))
            split['ToSt'].append(str(raw.loc[i[0],'ToSt']))
            split['Qty'].append(str(raw.loc[i[0],'Qty']))
            split['Line'].append(str(raw.loc[i[0],'Item #']))
            split['SLOC'].append(str(raw.loc[i[0],'Sloc']))
            split['Comment'].append(np.nan)
                
    except TypeError:
        continue
        
```

Output into the file you want. Copy the file path.

```python
df=pd.DataFrame(split)
df.to_excel("C:\\Users\\zhuoyan.bai\\Documents\\DO\\Line Split.xlsx",index=False)
```
